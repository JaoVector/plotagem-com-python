# Opções de pyplot.subplots

nrows - Número de Linhas das subplotagens \
ncols - Número de colunas das subplotagens \
sharex - Todas as subplotagens devem usar os mesmos tiques no eixo x (ajustar o xlim afetará todas as subplotagens) \
sharey - Todas as subplotagens devem usar os mesmos tiques no eixo y (ajustar o ylim afetará todas as subplotagens) \
subplot_kw - Dicionário de argumentos nomeados passados para a chamada a add_subplot para criar cada subplotagem. 

subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None) \

wspace e hspace - Controlam o percentual da largura e altura da figura

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500), bins=50, color="k", alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

# Cores Marcadores e Estilos de Linha

Os estilos de marcadores e linhas são passados para parametros dentro de plot como strings

```python
from numpy.random import randn
```

```python
plt.plot(randn(30).cumsum(), "ko--")
```

```python
# Outra forma de escrever a mesma expressão porém explicitamente
plt.plot(randn(30).cumsum(), color="k", linestyle="dashed", marker="o")
```

Quando utiliza-se a plotagen de linha, ocorre a interpolação linear entre pontos subsequentes, é possível utilizar o drawstyle para evitar esse problema.

```python
data = np.random.randn(30).cumsum()

plt.plot(data, "k--", label="Default")

plt.plot(data, "k-", drawstyle="steps-post", label="steps-post")

plt.legend(loc="best")
```
